```python
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from  nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
```

```python
df=pd.read_csv(r"C:\Users\VIRAJ\Downloads\bbc-text.csv")
df
```

```python
df["category"].value_counts()
```

```python
fv=df[["text"]]
cl=df[["category"]]
```

```python
fv.head()+
```

```python
stp=stopwords.words("english")
```

```python
new_df=[]
for y in df["text"]:
    l=[]
    z=y.lower()
    z=re.sub("[^a-z ]","",z)
    for word in word_tokenize(z):
        if word not in stp:
            nw=PorterStemmer().stem(word)
            l.append(nw)
    new_df.append(" ".join(l))
```

```python
fv=pd.DataFrame({"cleaned_text":new_df})
fv
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(fv,cl,train_size=0.8)
```

```python
x_train.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
dtm=cv.fit_transform(x_train["cleaned_text"])
dtm
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mnb=MultinomialNB()
```

```python
model=mnb.fit(dtm,y_train)
```

```python
model
```

```python
x_test
```

```python
x_test1=cv.transform(x_test["cleaned_text"])
```

```python
predict=model.predict(x_test1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predict)
```
